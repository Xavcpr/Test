# Analysis of Reviews on Olist

🎯 Now that you are familiar with NLP, let's analyze the reviews of Olist.

👇 Run the following cell to load the reviews dataset and install `unidecode`

```python
!pip install -q unidecode

import pandas as pd

url = "https://wagon-public-datasets.s3.amazonaws.com/Machine%20Learning%20Datasets/ml_olist_nlp_reviews.csv"
df = pd.read_csv(url, low_memory = False)

df.head()
```

```python
df.shape
```

```python
data = df[['length_review',  'review_score', 'review_comment_title', 'review_comment_message']]
data = data[data.length_review > 0]
data = data[data.review_score < 3].reset_index().drop(columns = 'index')
data
```

```python
data.review_comment_title[0] != 'nan'
```

```python
data.review_comment_title = data.review_comment_title.fillna(" ")
data.review_comment_message = data.review_comment_message + " " + data.review_comment_title
data
```

```python
import string
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python
def preprocessing(sentence):
    sentence = sentence.strip()
    sentence = sentence.lower()
    sentence = ''.join(char for char in sentence if not char.isdigit())
    for punctuation in string.punctuation:
        sentence = sentence.replace(punctuation, '')
    sentence = sentence.replace("  "," ")
#     tokenized_sentence = word_tokenize(sentence)
    word_tokens = word_tokenize(sentence)
    # Lemmatizing the verbs
    verb_lemmatized = [wnl.lemmatize(word, pos = "v") for word in word_tokens]
    # 2 - Lemmatizing the nouns
    noun_lemmatized = [wnl.lemmatize(word, pos = "n") for word in verb_lemmatized]
    final = ""
    for token in noun_lemmatized:
        final += token + " "
    return final
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
data['clean_reviews'] = data.review_comment_message.apply(preprocessing)
```

```python
df2=data.drop(columns = ['review_comment_title', 'review_comment_message', 'length_review'])
df2.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Vectorize the sentences
count_vectorizer_n_gram = CountVectorizer(ngram_range = (2,2)) # BI-GRAMS
review_vectorized_n_gram = count_vectorizer_n_gram.fit_transform(df2.clean_reviews)
```

```python
# Show the representations in a nice DataFrame
review_vectorized_n_gram = pd.DataFrame(
    review_vectorized_n_gram,
    columns = count_vectorizer_n_gram.get_feature_names_out(),
    index = df2.clean_reviews
)
# small = review_vectorized_n_gram[['producto errado']]
# Show the vectorized movies with bigrams
```
